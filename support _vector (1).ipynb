```python
from sklearn import preprocessing
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
forest["month"]=label_encoder.fit_transform(forest["month"])
```

```python
forest["day"]=label_encoder.fit_transform(forest["day"])
```

```python
forest["size_category"]=label_encoder.fit_transform(forest["size_category"])
```

```python
forest
```

```python
x=forest.iloc[:,:11]
```

```python
x
```

```python
y=forest["size_category"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import svm 
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
clf=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[50,5,10],'C':[15,14,13,12]}]
gsv=GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_,gsv.best_score_
```

```python
clf=SVC(C=15,gamma=50)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)*100
```

```python
print("Accurancy=",acc)
```

```python
confusion_matrix(y_test,y_pred)
```
